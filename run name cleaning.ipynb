```python
import pandas as pd
import regex as re
```

```python
df = pd.read_csv('data/ag2024runs_clean.csv')
```

```python
# removes the any% or 100% and following text from run names to make something closer to game names
def title_clean (x='txt'):
    if re.findall(r' \S+%.*', x) == []:
        return x
    else:
        return re.sub(r' \S+%.*','',x)
```
