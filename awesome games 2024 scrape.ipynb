```python
def URLtext(x = 1):
   return 'https://gamesdonequick.com/tracker/donations/AGDQ2024?page=' + str(x)
```

```python
URLtext(2)
```

```python
testlist = [i for i in range(1,3)]
testlist
```

```python
donlist1 = [i for i in range(1,351)]
```

```python
donlist2 = [i for i in range(351, 725)]
```

```python
donations = pd.DataFrame()
donations
```

```python
page_tables = []
for i in testlist:
    url = URLtext(i)
    response = requests.get(url)
    soup = BS(response.text)
    page_table = pd.read_html(str(soup.findAll('table')))[0]
    page_tables.append(page_table)
    donations = pd.concat(page_tables)
donations
```

```python
page_tables = []
for i in donlist1:
    print(i)
    url = URLtext(i)
    response = requests.get(url)
    soup = BS(response.text)
    page_table = pd.read_html(str(soup.findAll('table')))[0]
    page_tables.append(page_table)
donations1 = pd.concat(page_tables)
donations1
```

```python
page_tables2 = []
for i in donlist2:
    print(i)
    url = URLtext(i)
    response = requests.get(url)
    soup = BS(response.text)
    page_table = pd.read_html(str(soup.findAll('table')))[0]
    page_tables2.append(page_table)
donations2 = pd.concat(page_tables2)
donations2
```

```python
donations = pd.concat([donations1, donations2])
```

```python
donations = donations.reset_index(drop=True)
```

```python
donations.to_csv('data/ag2024donations.csv')
```
