```python
def URLtext(x = 1):
   return 'https://gamesdonequick.com/tracker/donations/SGDQ2023?page=' + str(x)
```

```python
#641 pages
```

```python
donlist1 = [i for i in range(1,320)]
```

```python
donlist2 = [i for i in range(320,642)]
```

```python
page_tables = []
for i in donlist1:
    print(i)
    url = URLtext(i)
    response = requests.get(url)
    soup = BS(response.text)
    page_table = pd.read_html(str(soup.findAll('table')))[0]
    page_tables.append(page_table)
donations1 = pd.concat(page_tables)
donations1
```

```python
page_tables2 = []
for i in donlist2:
    print(i)
    url = URLtext(i)
    response = requests.get(url)
    soup = BS(response.text)
    page_table = pd.read_html(str(soup.findAll('table')))[0]
    page_tables2.append(page_table)
donations2 = pd.concat(page_tables2)
donations2
```

```python
donations = pd.concat([donations1, donations2])
donations = donations.reset_index(drop=True)
```

```python
donations
```

```python
donations.to_csv('data/summer/donations.csv',index=False)
```

```python
runs = pd.read_csv('data/summer/runs.csv')
```

```python
runs.start = [pd.to_datetime(x[:-6]) for x in runs.start]
```

```python
runs.end = [pd.to_datetime(x[:-6]) for x in runs.end]
```

```python
donations = donations.rename(columns = {'Name':'name','Time Received':'time','Amount':'amount','Comment':'comment'})
```

```python
donations.time[0][:-6]
```

```python
times = [pd.to_datetime(x[:-6]) for x in donations.time]
```

```python
donations.time = times
```

```python
donations.amount = [x.replace('$','') for x in donations.amount]
```

```python
donations.amount = [x.replace(',','') for x in donations.amount]
```

```python
donations.amount = donations.amount.astype(float)
```

```python
run_totals = []
for ind, row in runs.iterrows():
    x = pd.Series(donations.loc[(donations['time'] > row['start']) & (donations['time'] < row['end'])]['amount']).sum().round(decimals=2)
    run_totals.append(x)
```

```python
runs['run_total'] = run_totals
```

```python
runs.sort_values('run_total').tail(10)
```
