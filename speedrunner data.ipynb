```python
runners_list = ['CaptainEzekiel', 'spicee', 'maeeeeee', 'gymnast86', 'Zic3', 'Froob']
```

response = requests.get('https://www.speedrun.com/api/v1/users/gymnast86/personal-bests?embed=game,category')

```python
entries = []
for ind in list(range(0,len(response.json()['data']))):
    entry = pd.DataFrame({
        'runner' : ['gymnast86'],
        'place' : [response.json()['data'][ind]['place']],
        'game' : [response.json()['data'][ind]['game']['data']['names']['twitch']],
        'category' : [response.json()['data'][ind]['category']['data']['name']],
        'time' : [response.json()['data'][ind]['run']['times']['primary_t']],
        'date' : [response.json()['data'][ind]['run']['date']]
    })
    entries.append(entry)
runs_df = pd.concat(entries)
runs_df
```

```python
def get_response(x):
    url = ('https://www.speedrun.com/api/v1/users/')+x+('/personal-bests?embed=game,category')
    
    response = requests.get(url)
    
    return response
```

```python
response = requests.get('https://www.speedrun.com/api/v1/users/spicee/personal-bests?embed=game,category')
response
```

```python
get_response('spicee')
```

```python
entries = []
for runner in runners_list:
    response = get_response(runner)
    
    for ind in list(range(0,len(response.json()['data']))):
        entry = pd.DataFrame({
            'runner' : [runner],
            'place' : [response.json()['data'][ind]['place']],
            'game' : [response.json()['data'][ind]['game']['data']['names']['twitch']],
            'category' : [response.json()['data'][ind]['category']['data']['name']],
            'time' : [response.json()['data'][ind]['run']['times']['primary_t']],
            'date' : [response.json()['data'][ind]['run']['date']]
        })
        entries.append(entry)
runs_df = pd.concat(entries)
runs_df
```

```python
time = pd.DataFrame(runs_df.groupby(by = 'runner')['time'].sum())
```

```python
time.time = time.time / 3600
```

```python
runner_stats = runner_stats.merge(time, on = 'runner')
```

```python
runner_stats.time = runner_stats.time.round(decimals=1)
```

```python
runner_stats = runner_stats.rename(columns = {'game_x': 'PB','game_y':'PB_Top3','game':'unique_games','pb2023':'PB_in_2023','time':'hours'})
```

```python
runner_stats = runner_stats.reset_index()
```
