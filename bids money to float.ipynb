```python
import pandas as pd
```

```python
bids = pd.read_csv('data/ag2024bids_clean.csv')
bids
```

```python
bids = bids.drop(columns = 'Unnamed: 0')
bids
```

```python
bids.amount = bids.amount.str.replace('$','').str.replace(',','')
```

```python
bids.goal = bids.goal.str.replace('—','0')
```

```python
bids.goal = bids.goal.str.replace('$','').str.replace(',','').astype(float)
```

```python
bids.to_csv('data/ag2024bids_clean.csv', index=False)
```
