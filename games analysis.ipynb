```python
dona.loc[dona.amount < 101].amount.plot.hist()
```

```python
tens_df = dona.loc[dona.amount == 10]
```

```python
run_totals = []
for ind, row in runs.iterrows():
    x = pd.Series(tens_df.loc[(tens_df['time'] > row['start']) & (tens_df['time'] < row['end'])]['amount']).sum().round(decimals=2)
    run_totals.append(x)
```

```python
runs['tens'] = run_totals
```

```python
runs[['name_x', 'tens']].sort_values('tens')
```

```python
dona['running_total'] = dona.amount.cumsum()
```

```python
dona_forwards = dona.iloc[::-1].reset_index(drop=True)
dona_forwards['running_total'] = dona_forwards.amount.cumsum()
```
